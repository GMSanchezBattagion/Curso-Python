```python
import os
import tamagotchi

# Limpia la consola
os.system('cls' if os.name == 'nt' else 'clear')

# Muestra el mensaje de bienvenida
print("Bienvenido al juego de Tamagotchi")
print("----------------------------------")
input("Presione Enter para comenzar...")
nombre=input( "Ingrese el nombre para su mascota: ")
# Crea una instancia de la clase Tamagotchi
mascota = tamagotchi.Tamagotchi(nombre)
# Bucle principal del juego
while True:
    # Limpia la consola
    os.system('cls' if os.name == 'nt' else 'clear')

    # Muestra el nombre de la mascota
    print(f"Nombre de la mascota: {mascota.nombre}")

    # Muestra las opciones al usuario
    print("Seleccione una opción:")
    print("1. Alimentar")
    print("2. Jugar")
    print("3. Dormir")
    print("4. Ver estado")
    print("5. Salir")

    # Obtiene la opción seleccionada por el usuario
    opcion = input("Ingrese una opción: ")

    # Realiza la acción seleccionada por el usuario
    if opcion == "1":
        mascota.alimentar()
    elif opcion == "2":
        mascota.jugar()
    elif opcion == "3":
        mascota.dormir()
    elif opcion == "4":
        mascota.mostrar_estado()
    elif opcion == "5":
        break
    else:
        print("Opción inválida")

    # Verifica el estado de la mascota
    mascota.verificar_estado()
```
