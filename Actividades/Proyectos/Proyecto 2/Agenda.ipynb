```python
"""Aplicación de Agenda de Contactos: Crea una aplicación de agenda de contactos que permita al usuario 
almacenar y gestionar información de contactos, como nombres, números de teléfono, direcciones de correo electrónico, etc.
El programa debe permitir agregar nuevos contactos, 
buscar contactos existentes (por diferentes criterios como número de teléfono, correo, nombre, etc),
actualizar información de contactos y 
eliminar contactos."""

```

```python
import json

def buscar_contacto(nombre_o_apellidos):
    with open('contactos.json', 'r') as file:
        agenda = json.load(file)
    
    contactos_encontrados = []
    
    for contacto in agenda['Contactos']:
        if (contacto['nombre'].lower().startswith(nombre_o_apellidos.lower()) or
            contacto['apellidos'].lower().startswith(nombre_o_apellidos.lower())):
            contactos_encontrados.append(contacto)
    
    if len(contactos_encontrados) == 0:
        print("No se encontraron contactos con los criterios de búsqueda.")
    else:
        for contacto in contactos_encontrados:
            print(f"ID:{contacto['id']}, Nombre: {contacto['nombre']} {contacto['apellidos']}, Email: {contacto['email']}, Teléfono: {contacto['telefono']}")

def mostrar_contactos():
    with open('contactos.json', 'r') as file:
        agenda = json.load(file)
        for contacto in agenda['Contactos']:
            print(f"ID:{contacto['id']} {contacto['nombre']} {contacto['apellidos']}, Email: {contacto['email']}, Teléfono: {contacto['telefono']}")

def agregar_contacto(nombre, apellidos, email, telefono):
    with open('contactos.json', 'r') as file:
        agenda = json.load(file)
    
    nuevo_contacto = {
        "id": len(agenda['Contactos']) + 1,
        "nombre": nombre,
        "apellidos": apellidos,
        "email": email,
        "telefono": telefono
    }
    
    agenda['Contactos'].append(nuevo_contacto)

    with open('contactos.json', 'w') as file:
        json.dump(agenda, file, indent=4)

def eliminar_contacto(id_contacto):
    with open('contactos.json', 'r') as file:
        agenda = json.load(file)
    
    encontrado = False
    
    for contacto in agenda['Contactos']:
        if contacto['id'] == id_contacto:
            agenda['Contactos'].remove(contacto)
            encontrado = True
            break
    
    if encontrado:
        with open('contactos.json', 'w') as file:
            json.dump(agenda, file, indent=4)
        print("Contacto eliminado correctamente.")
    else:
        print("No se encontró el contacto con el ID especificado.")

opciones = ""
print("Agenda de Contactos\n","1-Buscar contacto\n", "2-Agregar nuevo contacto\n", "3-Editar contacto existente\n", "4-Eliminar contacto\n", "5-Ver lista de contactos completa\n","6-Salir\n")
opciones = input("Seleccione la opcion que desee")
if opciones == "1":#buscar contacto
    nombre_o_apellidos = input("Ingrese el nombre o apellido del contacto a buscar")
    buscar_contacto(nombre_o_apellidos)
    pass
elif opciones == "2":#Agregar contacto
    apellidos= input("Ingrese el apellido")
    nombre=input("ingrese el nombre")
    email=input("ingrese el mail")
    telefono=input("ingrese el telefono")
    agregar_contacto(nombre, apellidos, email, telefono)
    print(nombre, apellidos, "fue agregado correctamente")
    input("aprete enter para continuar")
    pass
elif opciones == "3":#Editar
    pass
elif opciones == "4":#Eliminar
    nombre_o_apellidos=input("Ingrese el nombre o apellido del contacto a eliminar")
    buscar_contacto(nombre_o_apellidos)
    id_contacto = int(input("Para confirmar ingrese el ID del contacto a eliminar"))
    eliminar_contacto(id_contacto)
    pass
elif opciones == "5":#Ver lista de contactos completa
    mostrar_contactos()
    pass
elif opciones == "6":#salir
    pass
else:
    print("Seleccione una opcion válida")
    input("Aprete cualquier tecla para volver a empezar")
    pass
```
