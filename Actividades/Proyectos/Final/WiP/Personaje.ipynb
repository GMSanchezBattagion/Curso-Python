```python
# Clase base Personaje
class Personaje:
    def __init__(self, nombre, clase, nivel, vida, ataque, caracteristicas):
            self.nombre = nombre
            self.clase = clase
            self.nivel = nivel
            self.vida = vida
            self.ataque = ataque
            self.caracteristicas = caracteristicas

    def mostrar_caracteristicas(self):
        print(f"Características de {self.nombre}:")
        for caracteristica, valor in self.caracteristicas.items():
            print(f"{caracteristica}: {valor}")

    def atacar(self, objetivo):
        print(f"{self.nombre} ataca a {objetivo.nombre}!")
        objetivo.recibir_danio(self.ataque)

    def recibir_danio(self, cantidad):
        self.vida -= cantidad
        if self.vida <= 0:
            print(f"{self.nombre} ha sido derrotado.")
        else:
            print(f"{self.nombre} ha recibido {cantidad} puntos de daño.")    
    
    
    
#razas
# Elfos que hereda de Personaje
class Elfo(Personaje):
    def __init__(self, nombre, clase, nivel, vida, ataque, fuerza, destreza, constitucion, inteligencia, sabiduria, carisma):
        super().__init__(nombre, clase, nivel, vida, ataque, fuerza, destreza + 2, constitucion, inteligencia, sabiduria + 1, carisma)

#Enanos que hereda de Personaje
class Enano(Personaje):
    def __init__(self, nombre, clase, nivel, vida, ataque, fuerza, destreza, constitucion, inteligencia, sabiduria, carisma):
        super().__init__(nombre, clase, nivel, vida, ataque, fuerza + 2, destreza, constitucion + 2, inteligencia, sabiduria, carisma)

#Humanos que hereda de Personaje
class Humano(Personaje):
    def __init__(self, nombre, clase, nivel, vida, ataque, fuerza, destreza, constitucion, inteligencia, sabiduria, carisma):
        super().__init__(nombre, clase, nivel, vida, ataque , fuerza + 1, destreza + 1, constitucion + 1, inteligencia + 1, sabiduria + 1, carisma +1)
```
